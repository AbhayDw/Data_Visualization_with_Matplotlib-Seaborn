**some basic questions**

```python
#which team win mlost matches
match_wins= df["match_winner"].value_counts()
sns.barplot(y=match_wins.index,x=match_wins.values,palette='viridis')
plt.title("most match win by teams")
```

```python
#toss decision print
sns.countplot(x=df["toss_decision"],palette="magma")
```

```python
sns.countplot(x=df["toss_decision"],palette="crest")
plt.title("toss")
```

```python
#toss winner vs match winner
df["toss_match"]=df[df["toss_winner"]==df["match_winner"]]
sns.countplot(x="toss_match",data=df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df["toss_match_win"] = df["toss_winner"] == df["match_winner"]

sns.countplot(x="toss_match_win", data=df)

plt.xlabel("Toss Winner = Match Winner")
plt.ylabel("Count")
plt.title("Toss Impact on Match Result")

plt.show()
```

```python
df=pd.read_csv("apple.csv",usecols=["date","high","low","open","close"])
df
```

```python
df["date"]=pd.to_datetime(df["date"])
df.info()
```

```python
df.set_index("date",inplace=True)
```

```python
df
```

```python
df.loc["2016-05"].plot(y="close")
```

```python
#to change color simply use color
df.loc["2017-07"].plot(y=["open"],color="pink")
```

```python
df.loc["2017-05"].plot(y=["open"],color="red")
```

```python
#if you want a particular color so use hascode
```

```python
#to increase the line width
df.loc["2016-09"].plot(y=["open"],color="green",linewidth=5)
```

```python
#we also provide line style 4 type="solid___",dashed"----","dotted":",dashdot"-.
```

```python
from matplotlib.lines import lineStyles
df.loc["2016-08"].plot(y=["high"],color="orange",linewidth=3,linestyle="-")#solid
```

```python
from matplotlib.lines import lineStyles
df.loc["2016-08"].plot(y=["high"],color="orange",linewidth=3,linestyle="--")#dashded
```

```python
from matplotlib.lines import lineStyles
df.loc["2016-08"].plot(y=["high"],color="orange",linewidth=3,linestyle=":")#doted
```

```python
from matplotlib.lines import lineStyles
df.loc["2016-08"].plot(y=["high"],color="orange",linewidth=3,linestyle="-.")#dotdash
```
